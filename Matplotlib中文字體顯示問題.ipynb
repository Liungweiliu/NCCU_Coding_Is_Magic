## 方法二
* A.
    1. 到Windows系統字體資料夾中，找一個喜歡的中文字體檔案，e.g.simhei.ttf<br>
    2. 將Matplotlib中default的字體檔案Vera.ttf，所選的中文字體重新命名為Vera.ttf<br>

* B.
    1. 打開mpl-data資料夾中的matplotlibrc檔案，可以用Vim或記事本打開<br>

    >font.family:SimHei<br>
    >中間程式碼省略<br>
    >font.sans-serif:SimHei, Bitstream Vera Sans, Lucida Grande, Verdana...<br>

    2. 將axes.unicode_minus True改為False**<br>
    >axes.unicode_minus:False<br>

* C.
    修改rcsetup.py
    1. 打開rcsetup.py<br>
    
    >a map from key -> value, converter<br>>defaultParams = {<br>
    >省略<br>
    >'font.family':[[sans-serif], validate_stringlist]<br>
    >省略<br>
    >'font.sans-serif':[['SimHei','Bitstream Vera Sans', 'DejaVu Sans', 'Lucida Granda, 'Verdana'...]]<br>
    >}<br>

    2. 一樣將axes.unicode_minus True改為False<br>

    >a map from key -> value, converter<br>
    >defaultParams = {<br>
    >省略<br>
    >'axes.unicode_minus':[False, validate_bool],<br>
    >省略<br>
    >}<br>

### 完成A,B and C後就可以正常顯示中文了，不必額外執行程式碼

```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import IPython as ip
ip.display.set_matplotlib_formats('svg')
```

```python
#讀檔
df = pd.read_excel("TEJ_Excel練習_台積電.xlsx")
```

```python
#看columns
df.columns
```

```python
#切資料
closing_price = df.iloc[:,2]
opening_price = df.iloc[:,3]
```

```python
plt.figure(figsize=(10,7),dpi=100)
#分配上下兩圖 subplot(column row index)
p1 = plt.subplot(211)#(2columnwidth 1rowlength index1)
p2 = plt.subplot(212)#(2columnwidth 1rowlength index2)

#圖案主體
p1.plot(closing_price, color = "#F4247E", alpha = 0.6, lw=2.5)
p2.plot(opening_price, color = "#16A080", lw=2.5)

#背景格線
p1.grid(True, axis='y', ls='--', color='r', alpha=0.8)
p2.grid(True, axis='y', ls='--', color='g', alpha=0.8)

#正常顯示中、英文
p1.set_title('收盤價', color="#CE0059", fontsize = 20)
p2.set_title('$Opening$ $price$', color="#128469", fontsize = 20)

#y軸標籤
p1.set_ylabel('股價', color="#CE0059", fontsize = 20)
p2.set_ylabel('$Stock$ $Price$', color="#128469", fontsize = 20)
plt.savefig('Stock_price.png')
plt.savefig('Stock_price.svg')
plt.show()
```
